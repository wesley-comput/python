## Imóveis Residenciais

```python
import pandas as pd
```

```python
dados = pd.read_csv('aluguel.csv', sep = ';')
```

```python
dados.head(10)
```

```python
list(dados['Tipo'].drop_duplicates())
```

```python
residencial = ['Quitinete',
 'Casa',
 'Apartamento',
 'Casa de Condomínio',
 'Casa de Vila',]
```

```python
filtro = dados['Tipo'].isin(residencial)
```

```python
dados_residencial = dados[filtro]
```

```python
dados_residencial
```

```python
dados_residencial['Tipo'].drop_duplicates()
```

```python
dados_residencial.shape[0]
```

```python
dados_residencial.index = range(dados_residencial.shape[0])
```

```python
dados_residencial
```
