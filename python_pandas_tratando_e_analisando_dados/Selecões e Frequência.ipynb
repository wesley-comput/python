## Selecione somente os imóveis classificados com tipo 'Apartamento'

```python
filtro_apartamento = dados['Tipo'] == 'Apartamento'
```

```python
n1 = dados[filtro_apartamento].shape[0]
n1
```

## Selecione os imóveis classificados com tipos 'Casa', 'Casa de Condomínio' e 'Casa de Vila'.

```python
filtro2 = (dados['Tipo'] == 'Casa' ) | (dados['Tipo'] == 'Casa de Condomínio') | (dados['Tipo'] == 'Casa de Vila')
n2 = dados[filtro2]
n2.shape[0]
```

## Selecione os imóveis com área entre 60 e 100 metros quadrados, incluindo os limites.




```python
filtro3 = (dados['Area'] >= 60) & (dados['Area'] <= 100)
n3 = dados[filtro3]
n3
```

## Selecione os imóveis que tenham pelo menos 4 quartos e aluguel menor que R$ 2.000,00.

```python
filtro4 = (dados['Quartos'] >= 4) & (dados['Valor'] < 2000)
n4 = dados[filtro4]
n4.shape[0]
```
